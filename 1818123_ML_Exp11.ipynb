## Exp 11.Write a python program to implement Support Vector Machines classifier using credit card data set from kaggle.



```
import pandas as pd 
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn import svm
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import seaborn as sb
%matplotlib inline
```

```
df = pd.read_csv('creditcard.csv') 
df.head()
```

```
df_fraud = df[df['Class'] == 1]
plt.figure(figsize=(10,7))
plt.scatter(df_fraud['Time'], df_fraud['Amount'])
plt.title('Scatter plot amount fraud')
plt.xlabel('Time')
plt.ylabel('Amount')
plt.show()
```

```
df = df.dropna()
```

```
X = df.drop(['Time', 'Class'],axis=1) 
y = df['Class'] 
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```
df_corr = df.corr()
plt.figure(figsize=(15,10))
sb.heatmap(df_corr, cmap="YlGnBu")
plt.title('Heatmap correlation')
plt.show()
```

```
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, y_train)
```

```
X_test.dropna()
```

```
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
sb.heatmap(cm,annot=True,fmt = '')
```

```
accuracy_score(y_pred,y_test)
```
